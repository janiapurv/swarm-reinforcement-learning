### Example showing usage of RRT planner with Grid Map

```python
## Jupyter specific settings
%load_ext autoreload
%autoreload 2

import os
import sys
sys.path.insert(0, os.path.abspath('..'))
## End of Jupyter specific settings



import numpy as np

from planning.maps import GridObstacleMap
from planning.planners import RRT
from planning.plots import Plot3D
from planning.helpers import obstacle_generator, meshes


omap = GridObstacleMap( grid = obstacle_generator(np.full((100,100,100),255)) )

rrt = RRT( obstacle_map = omap
         , k    = 250
         , dt   = 1
         , init = (50,50,50)
         , low  = 0
         , high = 100
         , dim  = 3
         )
  
path = rrt.find_path( start = (0,0,0)
                    , goal  = (100,100,100))

Plot3D().draw(path, meshes=meshes)
```
